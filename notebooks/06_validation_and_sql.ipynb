```python
# 06_validation_and_sql

**Purpose:**  
Validate the end-to-end AI pipeline outputs and provide SQL queries  
used for dashboards and stakeholder analysis.

**Input Table:** logistics_gold.delivery_risk_decisions
```

```python
%python
final_df = spark.table("logistics_gold.delivery_risk_decisions")
```

```python
%sql
SELECT COUNT(*) AS total_shipments
FROM logistics_gold.delivery_risk_decisions;
```

```python
%sql
SELECT risk_category, COUNT(*) AS shipments
FROM logistics_gold.delivery_risk_decisions
GROUP BY risk_category;
```

```python
%sql
SELECT
  risk_category,
  ROUND(AVG(is_late), 3) AS actual_delay_rate
FROM logistics_gold.delivery_risk_decisions
GROUP BY risk_category;
```

```python
%sql
SELECT *
FROM logistics_gold.delivery_risk_decisions
WHERE risk_category = 'High'
ORDER BY delay_risk_score DESC
LIMIT 20;
```
